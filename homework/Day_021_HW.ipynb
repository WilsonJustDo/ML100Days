```python
# 題目超爛，寫得不清不楚，直接看解答要我們做什麼

# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np

# 準備數據 ... 設定要畫的的x,y數據list
x = np.arange(0, 3 * np.pi, 0.1)
x
```

```python
y_cos = np.cos(x)
y_cos
```

```python
# 畫出完整的 cos 圖形
plt.plot(x, y_cos)
plt.show()
```

```python
# 把 pyplot.show() 換成下面這行:
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

T = np.arctan2(Y,X)
print(T)
# 三角函數的arc tangent可以把X Y轉換成0~1的浮點數
# 剛好這些浮點數對應到scatter裡面的色碼錶
plt.scatter(X,Y, s=75, c=T, alpha=.5)
# s: 大小
# c: 顏色(這邊的顏色用arc tangent取得)
# alpha: 透明度
# marker: 形式

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
